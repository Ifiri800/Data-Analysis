Regrssion with  T-Test using Nitric Oxide concentrations and proportion of non-retail business acres per town

The hypothesis are: 

There is a significant difference in the median value of houses bounded by the Charles river or not

There is a difference in median values of houses of each proportion of owner-occupied units built before 1940 or
Can we conclude that there is no relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town?

There is impact of an additional weighted distance to the five Boston employment centres on the median value of owner-occupied homes

```python
## X is the input variables (or independent variables)
X = data['NOX']
## y is the target/dependent variable
y = data['MEDV']
## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

In Conclusion, like the T-Test, the P-value is less than the alpha level = 0.05, so, I reject the null hypothesis as there is evidence that there is a difference in mean evaluation score based on Nitix Oxide. The coefficient -33.9161 means that NOX get 0.339161 score less than MEDV.
